```python
import requests
from bs4 import BeautifulSoup
from openpyxl import Workbook
import re

# Import Modul
```

```python
# URL halaman yang ingin di-scrape
url = 'https://store.crunchyroll.com/collections/manga/'
```

```python
# Mengambil konten dari halaman web
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Mengambil semua item produk manga
manga_items = soup.find_all('div', class_='product')

# Memerika Isi Item
for item in manga_items :
    print(item)
```

```python
# Iterasi setiap item manga
for index, item in enumerate(manga_items, start=1):
    # Mencari nama produk dari tag <a> di dalam div
    product_name = item.find('a', class_='link').text.strip()
    
    # Mengambil slug produk dari href di dalam tag <a> (URL produk)
    product_link = item.find('a', class_='link')['href']
    product_slug = product_link.split('/')[-1].replace('.html', '')

    # Mencari Status Produk 
    status_product = item.find()
    
    # Mengambil harga sebelum diskon (price reduced from), jika ada
    price_before_tag = item.find('span', class_='strike-through')
    if price_before_tag:
        price_before = price_before_tag.find('span', class_='value').text.strip()
        pattern = r'\$\d+\.\d{2}'
        price = re.findall(pattern, price_before)
        prc = ''.join(price)
    else:
        price_before = "No previous price"
    
    # Mengambil harga setelah diskon (price after)
    price_after_tag = item.find('span', class_='sales')
    if price_after_tag:
        price_after = price_after_tag.find('span', class_='value').text.strip()
    else:
        price_after = "No sales price"

    # Mengambil status produk manga (PRE-ORDER atau lainnya)
    product_status_tag = item.find('div', class_='sash-content')
    if product_status_tag:
        product_status = product_status_tag.find('span').text.strip()
    else:
        product_status = "No status"
    
    # Output produk
    print(f"Product {index}:")
    print(f"Name: {product_name}")
    print(f"Slug: {product_slug}")
    print(f"Price Before: {prc}")
    print(f"Price After: {price_after}")
    print(f"Status Produk : {product_status}")
    print('-' * 40)
```
